```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df  = pd.read_csv('/Users/josedhernandez/Desktop/pseudo_facebook.csv')
df
```

```python
df.describe()
```

#  Cleaning the data

```python
# getting rid of N/A values
nan_df = df[df.isna().any(axis = 1)]
df= df.dropna()
df
# total of 177 rows with NaN values inside
```

# Relating Variables


```python
# Totaling all likes given and recieved into a new column
df['total given likes'] = df['likes'].astype('int')+df['mobile_likes'].astype('int')+df['www_likes'].astype('int')
df['total recieved likes'] = df['likes_received'].astype('int')+df['mobile_likes_received'].astype('int')+df['www_likes_received'].astype('int')
```

```python
GenderToLikes = df[['gender','total given likes','total recieved likes']]
GenderToLikes = GenderToLikes.dropna()
GenderToLikes
```

```python
sns.set()
_=plt.hist(df['age'])
_=plt.xlabel("Age Group")
_=plt.ylabel("Number of people in age group")
_=plt.title("People on facebook by age")
plt.show()
```

```python
sns.pairplot(data = GenderToLikes, hue="gender")
```

```python
# relation of likes given and likes recieved
sns.set()
ax = sns.regplot(x="total given likes", y="total recieved likes", data=GenderToLikes,x_estimator=np.mean)
```

```python
# Age and friendships initiated
sns.lmplot( x="age", y="friendships_initiated", data=df, fit_reg=False, hue='dob_year', legend=False)
plt.xlabel("Age")
plt.ylabel("Number of friendships initiated")
plt.title("Age & Friends initiated")
```
